## Using weighted average


>W=(Rv+Cm)/v+m



W=weighted Rating


R=Rating

v=votes

m=minimum voted required to be listed in the top 250(3000 movies)

C=the mean vote accross the whole report (6.9)

```python
v=merge_clean_df["vote_count"]
R=merge_clean_df["vote_average"]
C=merge_clean_df["vote_average"].mean()
m=merge_clean_df["vote_count"].quantile(0.70)
```

```python
merge_clean_df["Weighted_average"]=((R*v)+(C*m))/(v+m)
```

#Content Based Recommendation System


Recommendations based on movie's overview

```python
merge_clean_df["overview"].head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(
    min_df=3,
    max_features=None,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 3),
    stop_words='english'
)
```

```python
merge_clean_df["overview"]=merge_clean_df["overview"].fillna("")
```

```python
tfv_matrix=tfv.fit_transform(merge_clean_df["overview"])
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Suppose you want similar movies for movie index 5:
similarities = sig[10]  # Get similarities of movie 5 with all others
# Get top 5 similar movie indices (excluding itself)
similar_movies = similarities.argsort()[::-1][1:11]
print(similar_movies)
```

```python
indices = pd.Series(
    merge_clean_df.index,
    index=merge_clean_df['original_title']
).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return merge_clean_df['original_title'].iloc[movie_indices]
```

```python
give_rec("Interstellar")
```

```python
import ipywidgets as widgets
from IPython.display import display

# Define your recommend function
def recommend(title, sig=sig):
    idx = indices[title]
    sig_scores = list(enumerate(sig[idx]))
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    sig_scores = sig_scores[1:11]
    movie_indices = [i[0] for i in sig_scores]
    return merge_clean_df['original_title'].iloc[movie_indices]

# Create a text box widget
movie_title = widgets.Text(
    value='Avatar',
    placeholder='Type a movie title',
    description='Movie:',
    disabled=False
)

# Create an output area
output = widgets.Output()

# Define what happens when user presses Enter
def on_submit(change):
    with output:
        output.clear_output()
        try:
            recommendations = recommend(movie_title.value)
            print(f"Top recommendations for '{movie_title.value}':\n")
            for title in recommendations:
                print(title)
        except KeyError:
            print(f"Movie '{movie_title.value}' not found. Please check the title!")

# Bind the function
movie_title.on_submit(on_submit)

# Show the widget and output area
display(movie_title, output)
```
